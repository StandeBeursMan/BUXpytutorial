```python
import quandl
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

So, first of all, we need to import some modules. Don't forget to install them using pip. 

```python
data = quandl.get("WIKI/FB", end_date="2017-09-01")
print(data.tail())
```

We download Facebooks prices, WIKI is the quandl database of US stock data. We can use the .tail() function to inspect the last five datapoints. Next, we isolate the closeprices to make it simpler.

```python
closeprices = data['Close'] 
ma50 = closeprices.rolling(window=50).mean()
ma200 = closeprices.rolling(window=200).mean()
```

We use the .rolling and .mean functions to creat our moving averages. Then we plot them using matplotlibs .plot function and finally activate the graph with the .show function.

```python
plt.plot(closeprices)
plt.plot(ma50)
plt.plot(ma200)
plt.show()
```

If you run the plot function on your own system, the graph will be interactive and you can zoom in. If you want a shorter timeframe now, you choose a start date when downloading the data like so:

```python
data = quandl.get("WIKI/FB",start_date="2015-01-01", end_date="2017-09-01")
closeprices = data['Close'] 
ma50 = closeprices.rolling(window=50).mean()
ma200 = closeprices.rolling(window=200).mean()
plt.plot(closeprices)
plt.plot(ma50)
plt.plot(ma200)
plt.show()
```

Note that the 200 moving average has a 200 period delay before showing up.
